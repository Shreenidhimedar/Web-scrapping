```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.nykaa.com/makeup/face/c/13?ptype=lst&id=13&root=nav_2&dir=desc&order=popularity"
r=requests.get(url)
print(r)
```

```python
soup=BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
names=soup.find_all("div",class_="css-xrzmfa")
print(names)
```

```python
Product_name=[]
Mrp_price=[]
Review=[]
```

```python
names=soup.find_all("div",class_="css-xrzmfa")
for i in names:
    name=i.text
    Product_name.append(name)
    
print(Product_name)
```

```python
for i in names:
    name=i.text
    print(name)
```

```python
prices=soup.find_all("span",class_="css-111z9ua")
print(prices)
```

```python
for i in prices:
    mrp_price=i.text
    Mrp_price.append(mrp_price)
    
print(Mrp_price)
```

```python
reviews=soup.find_all("span",class_="css-1qbvrhp")
print(reviews)
```

```python
for i in reviews:
    review=i.text
    Review.append(review)
    
print(Review)
```

```python
print(len(Review))
```

```python
df =pd.DataFrame({"Product name":Product_name,"Mrp price": Mrp_price,"Review":Review})
print(df)
```
